```python
import pandas as pd

from ga_tools.optimization import GenericAlgorithm
```

```python
#　初期個体（なければcolumnsだけ定義)
cols = ['a', 'b','c', 'd', 'e']
df = pd.DataFrame(columns=cols)
ga = GenericAlgorithm(df)


def func(x):
    return sum(x)
weights = (1.0, )
ga.set_func([func], weights)


lower = [0, 0, 0, 0, 0]
upper = [1, 1, 1, 1, 1]
ga.set_lim(lower, upper)


n_population = 100
n_generation = 100
p_crossover = 0.5
p_mutation = 0.1
results = ga.run(n_population, n_generation, p_crossover, p_mutation)
```

## 結果の可視化
### 最適化されたx

```python
import matplotlib.pyplot as plt

plt.figure()
for result in results:
    plt.plot(result[:-1])
plt.ylim(0, 1)
plt.show()
```

```python
どの結果も1にちかく、one max problemが解かれていることがわかる
```

### 最適化されたy

```python
r_hist = [r[-1] for r in results]

plt.figure()
plt.hist(r_hist)
plt.show()
```
